```python
import pandas as pd
import matplotlib.pyplot as plt  # <-- necessário para usar plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# 2. Vendas por Categoria


```python
def contar_categorias(df, nome_loja):
    resumo = (
        df.groupby("Categoria do Produto")["Produto"]
        .count()
        .sort_values(ascending=False)
        .reset_index()
    )
    resumo.columns = ["Categoria", "Quantidade de Vendas"]
    print(f"\nTop categorias da {nome_loja}:")
    print(resumo.head(10))  # mostra as 10 mais populares
    return resumo

# Resultados por loja
cat_loja = contar_categorias(loja, "Loja 1")
cat_loja2 = contar_categorias(loja2, "Loja 2")
cat_loja3 = contar_categorias(loja3, "Loja 3")
cat_loja4 = contar_categorias(loja4, "Loja 4")
```

# 4. Produtos Mais e Menos Vendidos

```python
# --- Função para identificar produtos mais e menos vendidos ---
def produtos_populares(df, nome_loja):
    contagem = (
        df.groupby("Produto")["Produto"]
        .count()
        .sort_values(ascending=False)
        .reset_index(name="Quantidade de Vendas")
    )

    print(f"\n📈 {nome_loja} - Top 5 produtos mais vendidos:")
    print(contagem.head(5))
    print(f"\n📉 {nome_loja} - 5 produtos menos vendidos:")
    print(contagem.tail(5))

    # --- Visualização ---
    plt.figure(figsize=(8, 4))
    plt.barh(contagem.head(5)["Produto"], contagem.head(5)["Quantidade de Vendas"], color="seagreen")
    plt.gca().invert_yaxis()
    plt.title(f"{nome_loja} - Produtos mais vendidos")
    plt.xlabel("Quantidade de Vendas")
    plt.ylabel("Produto")
    plt.show()

    return contagem

# --- Executa para cada loja ---
vendas_loja = produtos_populares(loja, "Loja ")
vendas_loja2 = produtos_populares(loja2, "Loja 2")
vendas_loja3 = produtos_populares(loja3, "Loja 3")
vendas_loja4 = produtos_populares(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs dos CSVs
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
}

dfs = []

# Lendo os CSVs e adicionando a coluna 'loja'
for nome, url in urls.items():
    df_loja = pd.read_csv(url)
    df_loja['loja'] = nome
    dfs.append(df_loja)

# Unindo todos os DataFrames
df = pd.concat(dfs, ignore_index=True)

# Detectando automaticamente a coluna que contém 'frete' no nome
colunas_frete = [col for col in df.columns if 'frete' in col.lower()]
if not colunas_frete:
    raise ValueError("Não foi encontrada nenhuma coluna relacionada a frete.")
nome_frete = colunas_frete[0]  # pega a primeira coluna encontrada

print(f"Coluna usada para frete: {nome_frete}")

# Calculando custo médio de frete por loja
frete_medio_loja = df.groupby('loja')[nome_frete].mean().reset_index()
frete_medio_loja.columns = ['Loja', 'Custo Médio de Frete']

# Exibindo o resultado
print(frete_medio_loja)

# Gráfico
plt.figure(figsize=(8,5))
plt.bar(frete_medio_loja['Loja'], frete_medio_loja['Custo Médio de Frete'], color='skyblue')
plt.title('Custo Médio de Frete por Loja')
plt.xlabel('Loja')
plt.ylabel('Custo Médio de Frete (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

# 6. Gráficos


```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs das lojas
urls = {
    "Loja 1": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv",
    "Loja 2": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv",
    "Loja 3": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv",
    "Loja 4": "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
}

dfs = []
for nome, url in urls.items():
    df_loja = pd.read_csv(url)
    df_loja.columns = df_loja.columns.str.strip()  # remover espaços
    df_loja['loja'] = nome
    dfs.append(df_loja)

# Unindo todos os DataFrames
df = pd.concat(dfs, ignore_index=True)

# --- 1️⃣ Custo médio de frete por loja ---
frete_col = 'Frete'
frete_medio = df.groupby('loja')[frete_col].mean().reset_index()
frete_medio.columns = ['Loja', 'Custo Médio de Frete']

plt.figure(figsize=(8,5))
plt.bar(frete_medio['Loja'], frete_medio['Custo Médio de Frete'], color='skyblue')
plt.title('Custo Médio de Frete por Loja')
plt.xlabel('Loja')
plt.ylabel('Custo Médio de Frete (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- 2️⃣ Distribuição das categorias de produtos por loja (pizza) ---
cat_col = 'Categoria do Produto'

for loja_nome in df['loja'].unique():
    categorias = df[df['loja']==loja_nome][cat_col].value_counts()
    plt.figure(figsize=(7,7))
    plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
    plt.title(f'Distribuição das Categorias - {loja_nome}')
    plt.show()

# --- 3️⃣ Quantidade de produtos por categoria comparando lojas (barras horizontais) ---
quantidade_categoria = df.groupby(['loja', cat_col]).size().unstack(fill_value=0)

quantidade_categoria.plot(kind='barh', figsize=(10,6), stacked=True, colormap='tab20')
plt.title('Quantidade de Produtos por Categoria por Loja')
plt.xlabel('Quantidade de Produtos')
plt.ylabel('Loja')
plt.legend(title='Categoria', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# --- 4️⃣ Faturamento diário por loja (linha) ---
df['Data da Compra'] = pd.to_datetime(df['Data da Compra'])
valor_col = 'Preço'

for loja_nome in df['loja'].unique():
    faturamento_diario = df[df['loja']==loja_nome].groupby('Data da Compra')[valor_col].sum().sort_index()
    plt.figure(figsize=(10,5))
    plt.plot(faturamento_diario.index, faturamento_diario.values, marker='o', linestyle='-', label=loja_nome)
    plt.title(f'Faturamento Diário - {loja_nome}')
    plt.xlabel('Data')
    plt.ylabel('Faturamento (R$)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.show()

# --- 5️⃣ Avaliação dos clientes x preço da compra (dispersão) ---
avaliacao_col = 'Avaliação da compra'

for loja_nome in df['loja'].unique():
    plt.figure(figsize=(8,5))
    plt.scatter(df[df['loja']==loja_nome][valor_col], df[df['loja']==loja_nome][avaliacao_col],
                alpha=0.6, color='orange')
    plt.title(f'Avaliação dos Clientes x Preço da Compra - {loja_nome}')
    plt.xlabel('Preço da Compra (R$)')
    plt.ylabel('Avaliação do Cliente')
    plt.grid(True)
    plt.show()
```

# Relatório de Análise de Lojas

## **Introdução**
O objetivo desta análise é avaliar quatro lojas diferentes para determinar **qual delas o Senhor João deve vender**. Foram considerados diversos fatores, incluindo: faturamento total, categorias de produtos mais e menos vendidas, média das avaliações dos clientes, produtos mais e menos vendidos e custo médio de frete. A análise foi realizada utilizando os dados históricos de vendas de cada loja, e os resultados foram visualizados por meio de gráficos que facilitam a compreensão dos padrões e insights encontrados.

---

## **1. Faturamento Total das Lojas**
O faturamento total foi calculado somando o valor das vendas de cada loja. Observou-se que:  

- **Loja 1**: maior faturamento, com vendas consistentes ao longo do tempo.  
- **Loja 2**: faturamento médio, algumas flutuações ao longo das datas.  
- **Loja 3 e 4**: menor faturamento, com alguns dias de vendas muito baixas.  



**Insight:** lojas com maior faturamento tendem a ser mais atrativas para venda, pois indicam um histórico sólido de receitas.

---

## **2. Categorias de Produtos Mais e Menos Vendidas**
Analisando as categorias de produtos:  

- Cada loja possui produtos mais populares e outros com baixa saída.  
- **Loja 1** apresenta uma distribuição equilibrada entre categorias, com algumas categorias dominando as vendas.  
- **Loja 4** tem algumas categorias com baixa representatividade, indicando menor diversidade.  



**Insight:** lojas com diversidade de produtos vendidos indicam maior potencial de mercado, pois não dependem de poucas categorias.

---

## **3. Média das Avaliações dos Clientes**
A avaliação média dos clientes foi calculada para cada loja:  

- **Loja 1**: média alta, indicando satisfação do cliente consistente.  
- **Loja 2**: média razoável, algumas avaliações baixas.  
- **Loja 3 e 4**: médias mais baixas, com maior variabilidade.  



**Insight:** uma loja com avaliações consistentes e altas demonstra qualidade no atendimento e nos produtos, agregando valor à venda.

---

## **4. Produtos Mais e Menos Vendidos**
Os produtos mais vendidos são estratégicos para entender a força de cada loja:  

- **Loja 1**: apresenta produtos com alta rotatividade e vendas constantes.  
- **Loja 2**: alguns produtos vendem bem, mas outros ficam parados.  
- **Loja 3 e 4**: menos produtos de destaque, indicando menor potencial de receita.

---

## **5. Frete Médio**
O custo médio de frete foi analisado para cada loja:  

- **Loja 1**: frete médio competitivo, não impactando negativamente o custo total das vendas.  
- **Loja 2 e 3**: frete médio mais alto, podendo reduzir a margem de lucro.  
- **Loja 4**: frete variável, alguns dias com custo elevado.


**Insight:** frete mais baixo é um ponto positivo para o vendedor, reduzindo custos e tornando a loja mais atrativa.

---

## **Conclusão e Recomendação**

Após analisar os principais fatores: faturamento, diversidade de produtos, avaliação dos clientes, produtos vendidos e custo de frete, **a loja que se destaca é a Loja 1**.  

**Justificativas:**

1. **Maior faturamento total**, com vendas consistentes ao longo do tempo.  
2. **Diversidade de categorias e produtos vendidos**, diminuindo o risco de depender de poucos itens.  
3. **Avaliação dos clientes alta**, indicando satisfação e fidelização.  
4. **Frete médio competitivo**, preservando a margem de lucro.  

Portanto, recomenda-se que o Senhor João **priorize a venda da Loja 1**, pois ela apresenta **melhor desempenho financeiro, operacional e satisfação do cliente** em comparação às demais lojas.
